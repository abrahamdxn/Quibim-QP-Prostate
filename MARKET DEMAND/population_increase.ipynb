## Population Increase

```python
# Linear Regression Forecast for Population (10 years ahead)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
data = {
    "Year": list(range(2011, 2024)),
    "Population": [
        10543289, 10728970, 10902433, 11088234, 11250013,
        11423218, 11575567, 11704344, 11820426, 11885263,
        11967937, 12042776, 12109421
    ]
}
df = pd.DataFrame(data)

# Features (Year) and target (Population)
X = df["Year"].values.reshape(-1,1)
y = df["Population"].values

# Fit linear regression
linreg = LinearRegression().fit(X, y)

# Forecast future years (2024–2033)
future_years = np.arange(2024, 2034).reshape(-1,1)
lin_forecast = linreg.predict(future_years)

# Plot observed + forecast
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Population"], marker="o", label="Observed")
plt.plot(future_years, lin_forecast, "--", color="orange", label="Forecast (2024–2033)")
plt.title("Population Forecast (Linear Regression, 10 Years Ahead)")
plt.xlabel("Year"); plt.ylabel("Population")
plt.legend(); plt.show()

# Forecast table
forecast_df = pd.DataFrame({
    "Year": future_years.flatten(),
    "Forecast_Population": lin_forecast.astype(int)
})
display(forecast_df)
```

```python
# Linear Regression Forecast + Append 2023 & Forecasted 2024–2033
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data (2011–2023 actual)
data = {
    "Year": list(range(2011, 2024)),
    "Population": [
        10543289, 10728970, 10902433, 11088234, 11250013,
        11423218, 11575567, 11704344, 11820426, 11885263,
        11967937, 12042776, 12109421
    ]
}
df = pd.DataFrame(data)

# Features (Year) and target (Population)
X = df["Year"].values.reshape(-1,1)
y = df["Population"].values

# Fit linear regression model
linreg = LinearRegression().fit(X, y)

# Forecast future years (2024–2033)
future_years = np.arange(2024, 2034).reshape(-1,1)
lin_forecast = linreg.predict(future_years)

# Build forecast DataFrame
forecast_df = pd.DataFrame({
    "Year": future_years.flatten(),
    "Population": lin_forecast.astype(int)
})

# Append forecast to actuals
combined_df = pd.concat([df, forecast_df], ignore_index=True)

# Plot observed + forecast
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Population"], marker="o", label="Observed (2011–2023)")
plt.plot(forecast_df["Year"], forecast_df["Population"], "--", color="orange", label="Forecast (2024–2033)")
plt.plot(combined_df["Year"], combined_df["Population"], color="gray", alpha=0.5, label="Observed + Forecast Combined")
plt.title("Population Forecast (Linear Regression, 2011–2033)")
plt.xlabel("Year"); plt.ylabel("Population")
plt.legend(); plt.show()

# Show combined table
display(combined_df)
```
